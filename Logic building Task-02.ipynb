# Problem 1: Unique Words in a Sentence

```python
#input
sentence=input("Enter a sentence:").split()
# converted to set to remove duplicates
sentence=set(sentence)
print("Unique Words Count:",len(sentence))
print("Unique Words:",sentence)
```

# Problem 2: Highest Salary from Employee Data

```python
#input
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# iterating through dictionary to find highest salary
for employee,salary in employees.items():
    if  salary==max(employees.values()):
        print("Highest Salary:",employee,"-",salary)

```

# Problem 3: Find Maximum and Minimum Values

```python
numbers=list(map(int,input().split()))  # map is used to convert the string to int for every split
max=min=numbers[0]
for i in numbers[1:]:
    if i>max:
        max=i
    if i<min:
        min=i
print(numbers)
print("Max:",max)
print("Min:",min)
```

# Problem 4: Count Products Above a Price Threshold

```python
#input()
prices=input().split()
count=0
for i in prices:
    if int(i)>1000:    # string can be converted into integer using int
        count+=1
print("Products above 1000:",count)
```

# Problem 5: Calculate Attendance Percentage

```python
attendance=input().split()
present_days=attendance.count("P")
total_days=len(attendance)
print("Attendance Percentge:",(present_days/total_days)*100)
```

# Problem 7: Count Character Frequency

```python
string=input()
character_count=dict()
for i in string:
    character_count[i]=character_count.get(i,0)+1  # get() used to get the value of the key if doesnot exist writes deeafult key
print(character_count)
```
