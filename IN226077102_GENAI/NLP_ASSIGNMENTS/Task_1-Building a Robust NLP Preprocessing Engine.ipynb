# Task 2:Build Advanced Preprocessing Function

```python
import re
import string
def preprocess_text(text):

    if not isinstance(text,str):
        return []

    #Lowercasing
    text=text.lower()

    #removing links and mails
    text=re.sub(r'http\S+|www\S+','',text)

    #removing emails
    text=re.sub(r'S+@\S+','',text)

    #removing numbers
    text=re.sub(r'\d+','',text)

    #handling repeatated characters
    text=re.sub(r'(.)\1{2,}',r'\1',text)

    #removing punctuation
    text=text = text.translate(str.maketrans('', '', string.punctuation))

    #removing extra spaces
    text=re.sub(r'\s+',' ',text).strip()

    tokens=text.split()
    tokens=[token for token in tokens if len(token)>2 or token in ['no','not']]

    return tokens
    
```

# Task 3: Stress Testing

```python
inp=[
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in inp:
    cleaned_tokens=preprocess_text(text)
    cleaned_sentence=" ".join(cleaned_tokens)
    print("Original Text:",text,"\nCleaned Text:",cleaned_sentence,"\nCleaned Tokens:",cleaned_tokens)
    print()
    
```

# Task 4: Token Analytics

```python
for sentence in inp:
    tokens=preprocess_text(sentence)
    total_tokens=len(tokens)
    unique_tokens=len(set(tokens))
    avg_tokens=sum(len(w) for w in tokens)/total_tokens if tokens else 0

    print("Sentence:",sentence)
    print("Total Tokens:",total_tokens)
    print("Unique Tokens:",unique_tokens)
    print("Average Tokens:",avg_tokens)
    print()
```

# Task 5: Frequency Analysis

```python
from collections import Counter
all_tokens=[]

for sentence in inp:
    all_tokens.extend(preprocess_text(sentence))

counter=Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[:-6:-1])
```

# Task 6: Build Full Pipeline

```python
def full_pipeline(texts):
    tokens=[]
    clean_sentences=[]

    for text in texts:
        t=preprocess_text(text)
        tokens.extend(t)
        clean_sentences.append(" ".join(t))

    return {
        "tokens:":tokens,
        "clean_sentences:":clean_sentences
    }
result = full_pipeline(inp)
print(result)
```
