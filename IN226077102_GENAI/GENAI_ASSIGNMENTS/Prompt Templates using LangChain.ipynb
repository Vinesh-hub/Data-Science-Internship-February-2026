# Task 1: Replace Hardcoded Prompts (10%)


```python
from langchain_core.prompts import PromptTemplate

template=PromptTemplate(
    input_variables=["topic"],
    template='Explain {topic} in simple terms for beginners'
    )
result=template.format(topic="Machine Learning")
print(result)
```

# Task 2: Multi-Input Prompt System (15%)

```python
multi_template=PromptTemplate(
    input_variables=["topic","audience","tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)
test_cases=[
    {"topic":"Ai","audience":"beginners","tone":"friendly"}, #case-1
    {"topic":"Python","audience":"Kids","tone":"Fun"},       #case-2
    {"topic":"Deep Learning","audience":"engineers","tone":"technical"} #case-3
     ]
for test in test_cases:
    result=multi_template.format(**test)
    print(result)
```

# Task 4: ChatPromptTemplate System (15%)

```python
from langchain_core.prompts import ChatPromptTemplate
chat_template=ChatPromptTemplate.from_messages([
    ("system","You are a helpful {role}."),
    ("user","Explain {topic}")
])

chat_template.format_messages(
    role="teacher",
    topic="Neural Networks"
)
chat_template.format_messages(
    role="Interviewer",
    topic="tecnical"
)
chat_template.format_messages(
    role="Motivator",
    topic="Mindset"
)
```

# Task 5: Input Validation Layer (10%)

```python
def validate_inputs(audience,tone):
    if audience not in ["beginner", "intermediate","expert"]:
        print("Inavlid audience set to beginners")
        audience="beginner"
    if tone not in ["formal", "casual", "fun"]:
        print("Invalid tone set to formal")
        tone="formal"
    return audience,tone

audience=input("enter audience type:")
tone=input("Enter tone:")

audience,tone=validate_inputs(audience,tone)

print("Results:")
print("Audience:",audience,"\n","Tone:",tone)
```

# Task 6: Prompt Generator App (15%)

```python
def generate_prompt(topic,audience,tone,style):
    audience,tone=validate_inputs(audience,tone)
    template=PromptTemplate(
        input_variables=["topic","audience","tone","style"],
        template="Explain {topic} for {audience} in a {tone} {style} style"
    )
    return template.format(topic=topic,audience=audience,tone=tone,style=style)
print(generate_prompt(
    "Neaural Networks","beginner","fun","storytelling"
))
```
