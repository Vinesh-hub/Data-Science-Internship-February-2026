# Problem Statement 1:Smart Parking Lot Management System

```python
#function to manage a smart parking lot.
def parking_management(capacity,vehicle_logs):
    parked=0
    for i in vehicle_logs:
        if i.upper()=="IN":
            parked+=1
    print("Currently Parked Vehicle:",parked)
    print("Parking Status: Available" if capacity-parked>0 else "Parking Status: Not Available")

#input
capacity=int(input("Enter parking capacity"))
vehicle_logs=input("Enter vehicle log").split()

#function call
parking_management(capacity,vehicle_logs)


            
```

# Problem Statement 2:Online Food Delivery Time Estimator

```python
def deliverytime_estimator(distance,weather,traffic_level):
    # 5min for 1 km
    #if rainy 5 min delay
    # if traffic is high 10min,moderate 5min delay
    time=distance*5
    if weather.lower()=="rainy":
        time+=10
    if traffic_level.lower()=="high":
        time+=10
    elif traffic_level.lower()=="moderate":
        time+=5
    print("Estimated Delivary Time:",time,"minutes")

#input
distance=int(input("Distance (km)"))
weather=input("Weather:")
traffic_level=input("Traffic Level:")

#Function Call
deliverytime_estimator(distance,weather,traffic_level)
    
```

# Problem Statement 3:Movie Theatre Seat Occupancy

```python
def theatre_seat_occupancy(total_seats,booked_seats):
    occupied_percentage=(len(booked_seats)/total_seats)*100
    print("Occupancy:",occupied_percentage)
    if occupied_percentage==100:
        print("Show Status: HouseFull")
    elif occupied_percentage>=75:
         print("Show Status: Almost Full")
    elif occupied_percentage==50:
        print("Show Status: Half filled")
    else:
        print("Show Status: Seats available")

#input
total_seats=300
booked_seats=[1]*150

#function call
theatre_seat_occupancy(total_seats,booked_seats)
        
```

#  Problem Statement 4:Cloud Server Load Classification System

```python
def server_load_classify(cpu_readings):
    total=0
    for read in cpu_readings:
        total+=read
    avg_readings=total/len(cpu_readings)
    if avg_readings>=80:
        server_status="Critical"
    elif avg_readings>=50:
        server_status="Warning"
    else:
        server_status="Normal"
    print("Average CPU Load:",avg_readings)
    print("Server Status:",server_status)


# input
cpu_readings=[45, 60, 70, 85, 90]

# function call
server_load_classify(cpu_readings)               
```

#  Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def overuse_resourse(resource_usage):
    max_usage=max(resource_usage.values())
    for name,usage in resource_usage.items():
        if usage==max_usage:
            print("Overused Resources:",name)
            print("Energy Alert: Yes")

#input
resource_usage={
 "Projector": 6,
 "AC": 9,
 "Lights": 4
}

# function call

overuse_resourse(resource_usage)
```

#  Problem Statement 6: Online Event Registration Capacity Controller

```python
def capacity_controller(capacity,registrations):
    if registrations>event_capacity:
        waitlisted_users=registrations-event_capacity
        registration_status="Closed"
        confirmed_registrations=event_capacity
    else:
        waitlisted_users=0
        registration_status="Open"
        confirmed_registrations=registrations
    print("Waitlisted Users:",waitlisted_users)
    print("Registration Status:",registration_status)
    print("Confirmed Registrations:",confirmed_registrations)
# input
event_capacity=100
registrations=80

#function call
capacity_controller(capacity,registrations)
```
